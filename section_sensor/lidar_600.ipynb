```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt", delimiter=" ", header = None, names=("data", "time", "ir", "lidar"))
data["lidar"].hist(bins=max(data["lidar"]-min(data["lidar"])), align='left')
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"]=[e//10000 for e in data.time] # 
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = {i:d.lidar.get_group(i).value_counts().sort_index() for i in range(24)} #, each_hour
freqs = pd.concat(each_hour, axis=1) #concat
freqs = freqs.fillna(0) #(NaN)0
probs = freqs/len(data) #

probs
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum()) # 
p_t.plot()
p_t.transpose() # 
plt.show()
```

```python
p_t.sum() # 1
```

```python
p_z = pd.DataFrame(probs.transpose().sum())
p_z.plot()
plt.show()
p_z.transpose()
```

```python
# cond_z_t
cond_z_t = probs / p_t[0] #()P(t)P(x|t)
cond_z_t
```

```python
(cond_z_t[6]).plot.bar(color="blue", alpha=0.5)
(cond_z_t[14]).plot.bar(color="orange", alpha=0.5)
plt.show
```

```python
# p_z 
p_z
# p_z[0][0] -> 
# p_z[i][j] Pr(x1=i, x_2=j)
# p_z[0][0]
```

```python
# 
cond_t_z = probs.transpose()/probs.transpose().sum() # P(t|z)
# probs := P(z,t)
# p_z := P(z)
# p_t := P(t)

print("P(z=630)=", p_z[0][630]) # 630
print("P(t=13) =", p_t[0][13])  # 13
print("P(t=13|z=630)=",cond_t_z[630][13])
print("Bayes  P(z=630|t=13)=",cond_t_z[630][13]*p_z[0][630]/p_t[0][13]) # 2.51
print("answer P(z=630|t=13)=",cond_z_t[13][630]) # 13630
```

```python
def bayes_estimation(sensor_value,current_estimation):
    new_estimation = []
    for i in range(24):
        new_estimation.append(cond_z_t[i][sensor_value]*current_estimation[i])

    return new_estimation/sum(new_estimation) # 
```

```python
estimation = bayes_estimation(630,p_t[0])
plt.plot(estimation)
plt.show() # Live shareParticipants
# 
```

```python
value_5 = [630,632,636] # 5

estimation = p_t[0]
for v in value_5:
    estimation = bayes_estimation(v,estimation)

plt.plot(estimation)
plt.show() # 5
```
