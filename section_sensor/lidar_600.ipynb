```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt", delimiter=" ", header = None, names=("data", "time", "ir", "lidar"))
data["lidar"].hist(bins=max(data["lidar"]-min(data["lidar"])), align='left')
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"]=[e//10000 for e in data.time] # 切り捨て除算
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = {i:d.lidar.get_group(i).value_counts().sort_index() for i in range(24)} #時間ごとにデータフレームを作成, each_hourは辞書型の変数
freqs = pd.concat(each_hour, axis=1) #concatで連結
freqs = freqs.fillna(0) #欠損値(NaN)を0で埋める
probs = freqs/len(data) #頻度を確立で

probs
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum()) # 各列を合計
p_t.plot()
p_t.transpose() # 表示を横ベクトルにする
plt.show()
```

```python
p_t.sum() # 1になる
```

```python
p_z = pd.DataFrame(probs.transpose().sum())
p_z.plot()
plt.show()
p_z.transpose()
```

```python
# 条件付き確率cond_z_t
cond_z_t = probs / p_t[0] #列(時間)ごとにP(t)で割るとP(x|t)となる
cond_z_t
```

```python
(cond_z_t[6]).plot.bar(color="blue", alpha=0.5)
(cond_z_t[14]).plot.bar(color="orange", alpha=0.5)
plt.show
```

```python
# p_z は配列なのか確認
p_z
# p_z[0][0] -> データがないのでエラー
# p_z[i][j] は配列っぽく見えるが，Pr(x1=i, x_2=j)となる値を返すデータフレームである
# たとえばp_z[0][0]があるとは限らない
```

```python
# 条件付確率をベイズの定理から計算
cond_t_z = probs.transpose()/probs.transpose().sum() # 行と列を入れ替えて同様に計算するとP(t|z)となる
# probs := P(z,t)
# p_z := P(z)
# p_t := P(t)

print("P(z=630)=", p_z[0][630]) # センサ値が630になる確率（時間の情報は欠落）
print("P(t=13) =", p_t[0][13])  # 時間が13時である確率
print("P(t=13|z=630)=",cond_t_z[630][13])
print("Bayes  P(z=630|t=13)=",cond_t_z[630][13]*p_z[0][630]/p_t[0][13]) # 式2.51
print("answer P(z=630|t=13)=",cond_z_t[13][630]) # 13時にセンサ値が630
```

```python
def bayes_estimation(sensor_value,current_estimation):
    new_estimation = []
    for i in range(24):
        new_estimation.append(cond_z_t[i][sensor_value]*current_estimation[i])

    return new_estimation/sum(new_estimation) # 正規化
```

```python
estimation = bayes_estimation(630,p_t[0])
plt.plot(estimation)
plt.show() # これをしないとLive shareのParticipantsにはグラフが見えない
# 縦軸：確率、横軸：時間、
```

```python
value_5 = [630,632,636] # 得られたセンサ値（5時台のデータ）

estimation = p_t[0]
for v in value_5:
    estimation = bayes_estimation(v,estimation)

plt.plot(estimation)
plt.show() # グラフでも5時付近の確率が高くなっている
```

```python
value_11 = [617,624,619] # sensor_data_600.txtから拾ってきた11時台のセンサ値

estimation=p_t[0]
for v in value_11:
    estimation = bayes_estimation(v,estimation)

plt.plot(estimation)
plt.show()
```
