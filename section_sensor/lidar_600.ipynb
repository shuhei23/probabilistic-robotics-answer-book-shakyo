```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt", delimiter=" ", header = None, names=("data", "time", "ir", "lidar"))
data["lidar"].hist(bins=max(data["lidar"]-min(data["lidar"])), align='left')
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"]=[e//10000 for e in data.time] # 
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = {i:d.lidar.get_group(i).value_counts().sort_index() for i in range(24)} #, each_hour
freqs = pd.concat(each_hour, axis=1) #concat
freqs = freqs.fillna(0) #(NaN)0
probs = freqs/len(data) #

probs
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
sns.jointplot(data = data, x = "hour", y = "lidar", kind="kde")
plt.show()
```
